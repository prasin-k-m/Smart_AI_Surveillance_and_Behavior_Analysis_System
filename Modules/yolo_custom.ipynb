```python
import os
from ultralytics import YOLO
```

```python
def clean_split (image_dir, label_dir) :

    image_files = set(os.path.splitext(img)[0] for img in os.listdir(image_dir))
    label_files = set(os.path.splitext(label)[0] for label in os.listdir(label_dir))

    images_without_labels = image_files - label_files
    labels_without_images = label_files - image_files

    print("Images without labels:", images_without_labels)
    print("Labels without images:", labels_without_images)

    # Delete images without labels
    for name in images_without_labels:
        for ext in [".jpg", ".jpeg", ".png"]:
            img_path = os.path.join(image_dir, name + ext)
            if os.path.exists(img_path):
                os.remove(img_path)
                print("Deleted image:", img_path)

    # Delete labels without images
    for name in labels_without_images:
        lbl_path = os.path.join(label_dir, name + ".txt")
        if os.path.exists(lbl_path):
            os.remove(lbl_path)
            print("Deleted label:", lbl_path)

    print("Split cleaned!")
```

```python
# Clean training split
clean_split("dataset/images/train", "dataset/labels/train")
```

```python
# Clean validation split
clean_split("dataset/images/val", "dataset/labels/val")
```

```python
len(os.listdir("dataset/images/train")) == len(os.listdir("dataset/labels/train"))
len(os.listdir("dataset/images/val")) == len(os.listdir("dataset/labels/val"))
```
